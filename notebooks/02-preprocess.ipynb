# Set directories

```python
RAW_DIR = "../data/raw"
OUTPUT_DIR = "../data/processed/audio_16k"
TARGET_SR = 16000
TARGET_DURATION = 3.0
TARGET_LENGTH = int(TARGET_SR * TARGET_DURATION)

os.makedirs(OUTPUT_DIR, exist_ok=True)
```

# Process datasets

```python
rows = []

for dataset in os.listdir(RAW_DIR):
    dataset_path = os.path.join(RAW_DIR, dataset)
    if not os.path.isdir(dataset_path):
        continue
    
    for file in tqdm(os.listdir(dataset_path), desc=f"Processing {dataset}"):
        if file.lower().endswith((".wav", ".mp3", ".flac", ".ogg")):
            process_file(os.path.join(dataset_path, file), rows)
```

# Save metadata.csv

```python
df = pd.DataFrame(rows, columns=["file_path","emotion","dataset","actor", "index","sample_rate"])
df.to_csv("../data/processed/metadata.csv", index=False)
df.head()
```
